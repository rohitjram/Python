### Missing Value Handling

```python
null = ((df.isnull().sum())/len(df)).sort_values(ascending=False)
null[:60]
```

* **Dropping columns that have missing values of more than 30% as replacing them would affect our analysis**

```python
df.drop(null[null.values > 0.3].index, axis=1, inplace=True)
```

* Checking the missing values of remaining columns 

```python
df.isnull().sum().sort_values(ascending=False)[:20]
```

```python
df.DAYS_LAST_PHONE_CHANGE.plot.hist()
plt.show()
```

```python
df["AMT_ANNUITY"].plot.hist()
plt.show()
```

```python
df["AMT_GOODS_PRICE"].plot.hist()
plt.show()
```

* The values in the column are skewed. 
* So, we have to use median value to fill the missing values here.

```python
df["DAYS_LAST_PHONE_CHANGE"] = df["DAYS_LAST_PHONE_CHANGE"].fillna(df['DAYS_LAST_PHONE_CHANGE'].median())
df["CNT_FAM_MEMBERS"] = df["CNT_FAM_MEMBERS"].fillna(df["CNT_FAM_MEMBERS"].median())
df["AMT_ANNUITY"] = df["AMT_ANNUITY"].fillna(df['AMT_ANNUITY'].median())
```

* **Checking the rows that have 30% or more missing values in their respective rows**

```python
emptyrow=df.isnull().sum(axis=1)
emptyrow=list(emptyrow[emptyrow.values>=0.3*len(df)].index)
print(len(emptyrow))
```

* No row in the dataset has missing values of 30% or more

```python
df.CODE_GENDER.value_counts().plot.bar()
plt.show()
```

```python
df[df.CODE_GENDER == "XNA"]
```

* **Imputing the `XNA` values in the `CODE_GENDER` column with the most occuring value (`F`) in the column using mode as `XNA` is an unknown value. We do the imputation because there are only 4 records having `XNA` value in column `CODE_GENDER` and imputing these values will not affect our analysis.**

```python
df["CODE_GENDER"].replace("XNA", "F", inplace=True)
```

```python
df["ORGANIZATION_TYPE"].value_counts().sort_values(ascending=False)
```

* **Dropping the `XNA` values in the `ORGANIZATION_TYPE` column because it has a lot of missing values and imputing them can affect our analysis**

```python
df = df[~(df["ORGANIZATION_TYPE"] == "XNA")]
```

* **Defining functions for specific plots**

```python
def scatter(x, y, hue=None, title1=None, title2=None):
    plt.figure(figsize=(20,8))
    sns.set_style("whitegrid")
    plt.subplot(1,2,1)
    sns.scatterplot(x=target0[x],y=target0[y], hue=target0[hue])
    plt.title(title1)
    plt.subplot(1,2,2)
    sns.scatterplot(x=target1[x],y=target1[y], hue=target1[hue])
    plt.title(title2)
    plt.show()
    
    
def barplot(x, y, hue=None, title1=None, title2=None, figsize=None, rotation=None):
    plt.figure(figsize=(20,8))
    sns.set_style("whitegrid")
    plt.subplot(1,2,1)
    sns.barplot(x=target0[x],y=target0[y], hue=target0[hue])
    plt.title(title1)
    plt.xticks(rotation=rotation)
    plt.subplot(1,2,2)
    sns.barplot(x=target1[x],y=target1[y], hue=target1[hue])
    plt.title(title2)
    plt.xticks(rotation=rotation)
    plt.show()
    
def boxplot(x, y, hue=None, title1=None, title2=None):
    plt.figure(figsize=(20,8))
    sns.set_style("whitegrid")
    plt.subplot(1,2,1)
    sns.boxplot(x=target0[x],y=target0[y], hue=target0[hue])
    plt.title(title1)
    plt.subplot(1,2,2)
    sns.boxplot(x=target1[x],y=target1[y], hue=target1[hue])
    plt.title(title2)
    plt.show()
```
